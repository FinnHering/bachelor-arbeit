```python
from evaluate_benchmarks import parse_all_files
from pathlib import Path
import pandas as pd
```

```python
performances = parse_all_files(Path('apptainer'))
pfs = pd.concat(performances, ignore_index=True).reset_index()
df = pfs.groupby("name").agg({"operations": ["mean", "median", "std", "var", "skew"]})
df.sort_values(by=("name"), inplace=True, ascending=False)
pd.options.display.float_format = '{:.0f}'.format
df.to_excel("apptainer.xlsx")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path

# Plot histograms for each name on one figure
unique_names = pfs["name"].unique()
metrics = ["operations"]

fig, axes = plt.subplots(len(unique_names), figsize=(20 * len(metrics), 10 * len(unique_names)))

for i, name in enumerate(unique_names):
    name_res = pfs[pfs["name"] == name]
    sns.histplot(data=name_res, x="operations", ax=axes[i])
    axes[i].set_title(f"Histogram of operations for {name}")

plt.tight_layout()
plt.show()
```
